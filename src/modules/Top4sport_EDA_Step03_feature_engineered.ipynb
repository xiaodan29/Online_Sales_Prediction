```python
# List of columns to drop
columns_to_drop = ['total_sales_country_product', 'avg_sales_country_product', 'total_sales_country_product_month']

# Drop the specified columns
df1 = df1.drop(columns=columns_to_drop)
```

```python
# Calculate the total number of orders for each day
daily_orders = df1.groupby('date')['order_id'].nunique().reset_index()
daily_orders.columns = ['date', 'total_orders']
```

```python
# Calculate the total quantity of products sold for each day
daily_products = df1.groupby('date')['quantity'].sum().reset_index()
daily_products.columns = ['date', 'total_products']
```

```python
# Merge daily_orders and daily_products with the original dataset
df1 = df1.merge(daily_orders, on='date', how='left')
df1 = df1.merge(daily_products, on='date', how='left')

# Calculate the average number of orders per day
average_orders_per_day = daily_orders['total_orders'].mean()

# Calculate the average number of products sold per day
average_products_per_day = daily_products['total_products'].mean()

# Add these averages as new columns to the dataset
df1['average_orders_per_day'] = average_orders_per_day
df1['average_products_per_day'] = average_products_per_day
df1.head()
```

```python
# Sort the dataframe by 'date' and 'CreatedAt' columns
df1 = df1.sort_values(['CreatedAt'])
```

```python
# reset the index of the dataframe
df1 = df1.reset_index(drop=True)
```

```python
# Convert 'date' and 'CreatedAt' to datetime objects
df1['date'] = pd.to_datetime(df1['date'])

```

```python
# Group the dataframe by 'order_id', 'country', 'date'
grouped = df1.groupby(['order_id', 'date','country'])
```

```python
# Aggregate the categories and quantities for each group
aggregations = {
    'category': lambda x: x.value_counts().to_dict(),
    'payment': lambda x: x.value_counts().to_dict(),
    'gender': lambda x: x.value_counts().to_dict(),
    'age': lambda x: x.value_counts().to_dict(),
    'color': lambda x: x.value_counts().to_dict(),
    'size': lambda x: x.value_counts().to_dict(),
    'group0': lambda x: x.value_counts().to_dict(),
    'quantity': 'sum',
    'gift_quantity': 'sum' 
}

grouped_agg = grouped.agg(aggregations)
```

```python
# Calculate the total gross_revenue and profit for each group
revenue_profit_agg = grouped.agg({
    'gross_revenue': 'sum',
    'profit': 'sum'
})
```

```python
# Flatten the aggregated features and create a new dataframe
categories_agg_flat = grouped_agg.reset_index()
revenue_profit_agg_flat = revenue_profit_agg.reset_index()
```

```python
# Merge the flattened aggregated dataframes
date_related_df = pd.merge(categories_agg_flat, revenue_profit_agg_flat, on=['order_id', 'date', 'country'])
```

```python
additional_columns_df = df1[['order_id', 'price_difference', 'add_on_products', 'morning', 'afternoon',
                            'evening', 'avg_sales_country_product_day', 'total_orders','total_products', 
                            'average_orders_per_day', 'average_products_per_day']].copy()
```

```python
additional_columns_df = additional_columns_df.drop_duplicates(subset=['order_id'])
```

- Create new columns for unique categories, genders, and ages to make them simpler

```python
unique_categories = ['football', 'running', 'fitness','lifestyle']
unique_genders = ['men', 'women', 'unisex']
unique_ages = ['adults', 'kids']
unique_color = ['Red','White','Pink','Blue','Orange','Green','Black','Silver']
unique_group0 = ['Other','Apparel','Footwear','Equipment','Nutrition']

for cat in unique_categories:
    df_reset[f'category_{cat}'] = 0

for gen in unique_genders:
    df_reset[f'gender_{gen}'] = 0

for age in unique_ages:
    df_reset[f'age_{age}'] = 0
    
for color in unique_color:
    df_reset[f'color_{color}'] = 0
    
for group0 in unique_group0:
    df_reset[f'group0_{group0}'] = 0

# Set the values based on the original data
for index, row in df_reset.iterrows():
    # For categories
    for cat in unique_categories:
        if cat in row['category']:
            df_reset.at[index, f'category_{cat}'] = 1

    # For genders
    for gen in unique_genders:
        if gen in row['gender']:
            df_reset.at[index, f'gender_{gen}'] = 1

    # For ages
    for age in unique_ages:
        if age in row['age']:
            df_reset.at[index, f'age_{age}'] = 1

    # For colors
    for color in unique_color:
        if color in row['color']:
            df_reset.at[index, f'color_{color}'] = 1
            
    # For sizes
    for group0 in unique_group0:
        if group0 in row['group0']:
            df_reset.at[index, f'group0_{group0}'] = 1
# Drop the original columns
df_reset.drop(['category', 'gender', 'age','color','group0','size'], axis=1, inplace=True)
```

```python
unique_payment = ['COD','GOPAY_CARD','PAYPAL','CARD_GOPAY','CARD_CSOB','CHECKOUT_CARD','PAYPAL_EXPRESS',
                  'CHECKOUT_APPLE','CHECKOUT_BANCONTACT','BANK_TRANSFER','CASH','CSOB_CARD','INVOICE_HUF','PAYPAL_IDEALO']

for payment in unique_payment:
    df_reset[f'payment_{payment}'] = 0

for index, row in df_reset.iterrows():
    # For payment
    for payment in unique_payment:
        if payment in row['payment']:
            df_reset.at[index, f'payment_{payment}'] = 1

# Drop the original columns
df_reset.drop(['payment'], axis=1, inplace=True)
```
